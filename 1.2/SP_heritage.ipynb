```python
# Introducción 

"""Este proyecto se trata de un trabajo de corpus sobre las diferencias lingüísticas entre hablantes de herencia, hablantes 
nativos y hablantes de segunda lengua"""


## Adquisición de los dados para los hablantes nativos
import requests
import zipfile
import StringIO

zip_file_url = 'http://lenguajeacademico.info/proyecto/corpus_CLAE.zip'
r = requests.get(zip_file_url)
z = zipfile.ZipFile(StringIO.StringIO(r.content))
z.extractall()

%cd corpus_CLAE/
%rm C*.txt
%ls
```

```python
## Construcción del corpus y las distribuciónes según la frequencia del vocabulario, con las palabras comunes
from nltk.corpus import PlaintextCorpusReader, stopwords
from nltk.text import Text
from nltk.probability import FreqDist

corpus_root = '/Users/christopherstewart/Desktop/sp_heritage/1.1/corpus_CLAE'
corp = [w.lower() for w in Text(PlaintextCorpusReader(corpus_root, '.*', encoding='utf-8').words()) if w.isalpha()]
corp_fd = FreqDist(corp)
corp_fd.tabulate(20)
```

```python
## Y ahora sin las palabras comunes

stopset = set(stopwords.words('spanish'))
corp_2 = [w for w in corp if w not in stopset]
corp_2_fd = FreqDist(corp_2)
corp_2_fd.tabulate(15)
```

```python
## Aislamiento de colocaciones de dos palabras CON las palabras más comunes
from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures

### unicode already messed up here
bcf = BigramCollocationFinder.from_words(corp)
bcf.nbest(BigramAssocMeasures.likelihood_ratio, 100)
```

```python
## Aislamiento de colocaciones de dos palabras SIN las palabras más comunes
from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures

bcf = BigramCollocationFinder.from_words(corp_2)
bcf.nbest(BigramAssocMeasures.likelihood_ratio, 100)
```

```python
## Aislamiento de colocaciones de dos palabras SIN las palabras más comunes
from nltk.util import bigrams

bigrs = FreqDist(bigrams(corp_2))
bigrs.most_common(100)
```

```python
## Aislamiento de colocaciones de tres palabras SIN las palabras más comunes
from nltk.collocations import TrigramCollocationFinder
from nltk.metrics import TrigramAssocMeasures

tcf = TrigramCollocationFinder.from_words(corp_2)
tcf.nbest(TrigramAssocMeasures.likelihood_ratio, 20)
```

```python
## Aislamiento de colocaciones de dos palabras SIN las palabras más comunes
from nltk.util import trigrams

trigrs = FreqDist(trigrams(corp))
trigrs.most_common(20)
```

```python
from nltk.util import ngrams

tetragrs = FreqDist(list(ngrams(corp, 4)))
tetragrs.most_common(20)
```

```python
pentagrs = FreqDist(list(ngrams(corp, 5)))
pentagrs.most_common(20)
```
